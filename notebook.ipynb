Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a maximum score of 800 points. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()

# Start coding here...
# Add as many cells as you like...
```

```python
# Quick look on df structure
schools.info()
```

```python
# 1st Question
# Finding max average math score
max_math = schools['average_math'].max()

# Filtering only schools where results are at least 80% of max
high_scores = schools[schools['average_math'] >= 640]

# Final answer
best_math_schools = high_scores[['school_name', 'average_math']].sort_values(by = 'average_math', ascending=False)
best_math_schools.head()
```

```python
# 2d Question
# Add a total column
schools['total_SAT'] = schools['average_math'] + schools['average_reading'] + schools['average_writing']

# Final answer, top 10 schools
top_10_schools = schools[['school_name', 'total_SAT']].sort_values(by = 'total_SAT', ascending=False).head(10)
print(top_10_schools)
```

```python
# 3d Question
# Object: locate a borough w/ largest SD for total_SAT
# 1 step - Add column w/ SD for total_sat
schools['SD_for_total_SAT'] = schools['total_SAT'].std()

# Count schools in borough
schools_in_borough = schools.groupby('borough').agg(
                        num_schools = pd.NamedAgg(column = 'school_name', aggfunc='count'),
                        average_SAT = pd.NamedAgg(column = 'total_SAT', aggfunc='mean'),
                        std_SAT = pd.NamedAgg(column = 'total_SAT', aggfunc='std'))
    
# Round df
schools_in_borough_rounded = schools_in_borough.round(2)

# Answer
largest_std_dev = schools_in_borough_rounded.sort_values(by = 'std_SAT', ascending=False).head(1)
print(largest_std_dev)
```
